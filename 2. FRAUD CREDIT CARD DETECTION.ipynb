# 4. UNDERSAMPLING TECHNIQUES FOR IMBALANCED DATASET

## Here are the 3 of the most common UnderSampling Methods when dealing with Imbalanced Dataset

1. RandomUnderSampler - Randomly removes samples from the majority class until balance is achieved.
2. SMOTE (Synthetic Minority Oversampling Technique) - Generates synthetic samples for the minority class by interpolating between neighbors.
3. SMOTE‑Tomek - Combines SMOTE oversampling with Tomek links (removes borderline samples where classes overlap).

## Practical Applications
- If your dataset is large → start with RandomUnderSampler (quick, efficient).
- If your dataset is small or minority class is tiny → use SMOTE to generate synthetic samples.
- If your dataset has class overlap/noisy boundaries → use SMOTE‑Tomek for both balancing and cleaning.

On this dataset which is large - 284 thousand rows, we'll use SMOTE-Tomek Undersampling for both balanced result

# 5. TRAINING MODEL WITHOUT UNDERSAMPLING

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% test, 80% train
    stratify=y,           # preserve class ratios
    random_state=random_state
)

X_train = X_train.to_numpy()
y_train = y_train.to_numpy().flatten()
X_test = X_test.to_numpy()
y_test = y_test.to_numpy().ravel()
```

```python
models = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "AdaBoostClassifier": AdaBoostClassifier()
}
```

```python
results = {
    "Model": [],
    "Accuracy": [],
    "F1 Score Weighted": [],
    "F1 Score Macro": [],
    "Precision_weighted": [],
    "Recall_weighted": [],
    "Precision_macro_avg": [],
    "Recall_macro_avg": []
}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    results["Model"].append(model_name)
    results["Accuracy"].append(accuracy_score(y_test, y_pred))
    results["F1 Score Weighted"].append(f1_score(y_test, y_pred, average='weighted'))
    results["F1 Score Macro"].append(f1_score(y_test, y_pred, average='macro'))
    results["Precision_weighted"].append(precision_score(y_test, y_pred, average='weighted'))
    results["Recall_weighted"].append(recall_score(y_test, y_pred, average='weighted'))
    results["Precision_macro_avg"].append(precision_score(y_test, y_pred, average='macro'))
    results["Recall_macro_avg"].append(recall_score(y_test, y_pred, average='macro'))
```

### RESULT W/OUT UNDERSAMPLING

```python
results_df = pd.DataFrame(results)
print(results_df)
```

# 6. TRAINING MODEL WITH UNDERSAMPLING

```python
smote = SMOTETomek(random_state=random_state)

X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

print("X_train Before SMOTETomek: \n", pd.Series(y_train).value_counts())
print("X_train After SMOTETomek: \n", pd.Series(y_resampled).value_counts())

print("y_train Before SMOTETomek: \n", pd.Series(y_train).value_counts())
print("y_train After SMOTETomek: \n", pd.Series(y_resampled).value_counts())
```

```python
undersampled_models = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "AdaBoostClassifier": AdaBoostClassifier()
}
```

```python
undersampled_results = {
    "Model": [],
    "Accuracy": [],
    "F1 Score Weighted": [],
    "F1 Score Macro": [],
    "Precision_weighted": [],
    "Recall_weighted": [],
    "Precision_macro_avg": [],
    "Recall_macro_avg": []
}

# Train and evaluate
for model_name, model in undersampled_models.items():
    model.fit(X_resampled, y_resampled)
    y_pred = model.predict(X_test)
    
    undersampled_results["Model"].append(model_name)
    undersampled_results["Accuracy"].append(accuracy_score(y_test, y_pred))
    undersampled_results["F1 Score Weighted"].append(f1_score(y_test, y_pred, average='weighted'))
    undersampled_results["F1 Score Macro"].append(f1_score(y_test, y_pred, average='macro'))
    undersampled_results["Precision_weighted"].append(precision_score(y_test, y_pred, average='weighted'))
    undersampled_results["Recall_weighted"].append(recall_score(y_test, y_pred, average='weighted'))
    undersampled_results["Precision_macro_avg"].append(precision_score(y_test, y_pred, average='macro'))
    undersampled_results["Recall_macro_avg"].append(recall_score(y_test, y_pred, average='macro'))
```

### RESULT W/ UNDERSAMPLING

```python
results_df = pd.DataFrame(undersampled_results)
print(results_df)
```

Comparing Results:

1. Classification without undersampling  
The metrics appeared extremely high (close to 100%). However, the model was biased toward the majority class. This can look “perfect,” but in reality minority detection was weak and the model risked overfitting.

3. Classification with undersampling applied (SMOTE‑Tomek)

In domains such as fraud detection, this trade‑off is valuable, since correctly identifying minority cases is more important than inflated overall accuracy.

We can use the undersampled K-Nearest Neighbors (KNN) classifier since based on the evaluation metrics, it achieved the highest performance among all models. 
Going back to data description. The dataset has already been transformed using PCA, which enhances separability and contributes to stronger results. KNN is particularly effective when working with clustered datasets, making it a suitable choice here.

Weighing accuracy vs F1 and precision/recall is critical.

It depends on your domain in acheiving business goal.

a. Accuracy - Good for overall performance, but can mislead if dataset is imbalanced (like our dataset).
b. F1 Score - Harmonic mean of precision and recall. It balances precision and recall in a single metric.
c. Precision - Out of all the samples the model predicted as positive, how many were actually positive? It avoids false alarms.
d. Recall - Out of all the actual positive samples, how many did the model correctly identify? Goal is to catch all frauds including possible non fraud transaction.

e. Weighted - Good when overall accuracy matters, but can hide minority class weakness.
f. Macro - Best for imbalanced datasets where minority detection is critical.
